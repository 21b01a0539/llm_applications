```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader(
    "example.pdf",
)
```

```python
pdf = loader.load()
type(loader.load()[0])
```

```python
from langchain_community.document_loaders import WebBaseLoader
loader = WebBaseLoader(
    "https://en.wikipedia.org/wiki/Python_(programming_language)"
)
```

```python
docs = loader.load()
docs[0]
```

Text splitters split documents into smaller chunks for use in downstream applications.

Different types of text-splitters 
Lenght-based (Character text splitter)
Text-structure based (Recursive character text splitter)
Document-structured based (Markdown,HTML,JSON,Code etc...)

```python
##text based chuncking
from langchain.text_splitter import CharacterTextSplitter
text_splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap=0)
docs = text_splitter.split_documents(pdf)
docs[0]
```

```python
##recursive text splitter
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=0)
docs = text_splitter.split_documents(pdf)
docs
```
